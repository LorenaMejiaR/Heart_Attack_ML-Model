```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
heart_data = "../data/raw/heart_disease_health_indicators_BRFSS2015.csv"
heart_df = pd.read_csv(heart_data)
```

```python
correlation_matrix = heart_df.corr()

# Setting up the matplotlib figure with an appropriate size
plt.figure(figsize=(18, 15))

# Drawing the heatmap for the numerical columns
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap for Selected Numerical Variables")
plt.show()
```

```python
target = heart_df['HeartDiseaseorAttack']
num_df = heart_df.drop('HeartDiseaseorAttack', axis=1)

x_train, x_test, y_train, y_test = train_test_split(num_df, target, test_size=0.20, random_state=0)
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
print(f"The accuracy of the model vs the test is {knn.score(x_test, y_test)*100: .2f}%")
print(f"The accuracy of the model vs the training data is {knn.score(x_train, y_train)*100: .2f}%")

```
